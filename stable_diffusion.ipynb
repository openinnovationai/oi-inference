### OICM Inference for models deployed with: OI SERVE
##### text-to-image

```python
import os
from openai import OpenAI
import base64
from PIL import Image
import io
from IPython.display import display
```

```python
api_key = os.getenv('OI_API_KEY') # Your valid api key
deployment_id = os.getenv('OI_DEPLOYMENT_ID') # deployment id
base_url = f"https://inference.stg.openinnovation.ai/models/{deployment_id}/proxy/v1" # replace "stg" with your environment
model = "stabilityai/stable-diffusion-3.5-medium" # replace tje model name with your model name
```

#### Call the model 

```python
client = OpenAI(
    base_url=base_url,
    api_key=api_key
)
```

```python
img = client.images.generate(
    model=model,
    prompt="A majestic falcon statue crafted from polished gold, highly detailed, intricate feathers, glowing in warm ambient light, placed on a marble pedestal, dramatic shadows, hyper-realistic.",
)
```

```python
image_bytes = base64.b64decode(img.data[0].b64_json)
image = Image.open(io.BytesIO(image_bytes))
display(image)
```
